```python
def split_dataset(dataset, split_ratio=0.8):
    """Permet de separer les données en jeux d'entrainement et de validation selon un pourcentage pour le jeu d'entrainement"""

    random.shuffle(dataset)  # Mélangez les données aléatoirement

    split_index = int(len(dataset) * split_ratio)

    train_set = dataset[:split_index]
    validation_set = dataset[split_index:]

    return train_set, validation_set


train_set, validation_set = split_dataset(np.array(df_train), split_ratio=0.8)

#Division des labels et des données
X_train = train_set[:,:-1]
Y_train = train_set[::,-1]
X_val = validation_set[:,:-1]
Y_val = validation_set[::,-1]


print("=============== NORMALISATION ==============")

def normalisation(data):
    moyenne = data.mean(axis=0)
    ecart_type = data.std(axis=0)

    # Normalisez les colonnes sélectionnées (Soustraction de la moyenne et division par l'écart type)
    data = (data - moyenne) / ecart_type

    return data

X_train[:,1:] = normalisation(X_train[:,1:])

colone = df_train.columns
print(colone[:-1])
df3 = pd.DataFrame(X_train, columns=colone[:-1])
df3
```
